### 3.1 Removing Header Words <a class="anchor" id="subheader31"></a>

We'll first check to see if these paragraphs do contain the information we expect.

### 3.2 Stratifying the Data <a class="anchor" id="subheader32"></a>

Given that we can't simply stratefy our train, validation and test splits on a continuous target variable, adding a categorical value corresponding to `overall_score`. This allows train, val, test to be stratefied, allowing for even distribution of values for training, validation and testing. We'll drop our temporary extra y column before running any tests.

We're going to use a bit more robust mapper in this notebook than the last because of how many scores are clustered in the 80 and 90s.

```python
#plot y distribution
plt.figure()
plt.hist(coffee['overall_score'])
plt.title('historgram of overall_score')
plt.show()
```

```python
#creating categorical mapper
score_mapper = {range(0,70):'one', range(70,80):'two', range(80,85):'three', range(85,90):'four', range(90,95):'five', range(95,101): 'six'}
```

```python
#applying the categorical mapper
coffee['overall_cat'] = coffee['overall_score'].replace(score_mapper)
```

```python
#verifying splits appear accurate
coffee[['overall_score', 'overall_cat']].sample(20)
```

### 4.1 CountVectorizer() <a class="anchor" id="subheader41"></a>

CountVectorizer() has a number of parameters that can be adjusted. The following uses a pipeline and gridsearch to optimize these parameters. To find the optimal parameters, different parameter ranges were tested, starting with more extreme differences and narrowing down until it arrived at the best results. The cell showing 'parameters' now only shows the final results.

**Optimizing CountVectorizer with Linear Regression** <a class="anchor" id="sub2header411"></a>

*`p1` optimization using Linear Regression:*

```python
#creating our pipeline
#the values listed here for vectorization are placeholders
linreg = Pipeline(
    [
        ("vectorizer", CountVectorizer(stop_words = 'english', min_df=1, max_df=5000, max_features=2000)),
        ("classifier", LinearRegression()),
    ]
)
```

```python
#ran the grid search using a series of changes on these paramaters
#started with more extreme options for each paramater and then narrowed down to these

parameters = {
    "vectorizer__min_df": [9], 
    "vectorizer__max_df": [3000], 
    "vectorizer__max_features": [650],
    "vectorizer__stop_words": ['english'],
}
```

```python
#running the serach to find the best combination of these parameters
grid_search = GridSearchCV(linreg, parameters)

grid_search.fit(X_remain['p1'], y_remain)

print("Best parameters: ", grid_search.best_params_)
```

```python
#checking the R2
grid_search.best_score_
```

*`p2` optimization using Linear Regression:*

```python
#creating our pipeline
#the values listed here for vectorization are placeholders and will be replaced when we run the gridsearch
linreg = Pipeline(
    [
        ("vectorizer", CountVectorizer(stop_words = 'english', min_df=1, max_df=5000, max_features=2000)),
        ("classifier", LinearRegression()),
    ]
)

#ran the grid search using a series of changes on these paramaters
#started with more extreme options for each paramater and then narrowed down to these 3

parameters = {
    "vectorizer__min_df": [12], 
    "vectorizer__max_df": [3100], 
    "vectorizer__max_features": [500],
    "vectorizer__stop_words": ['english'],
}
```

```python
#running the serach to find the best combination of these parameters
grid_search = GridSearchCV(linreg, parameters)

grid_search.fit(X_remain['p2'], y_remain)

print("Best parameters: ", grid_search.best_params_)

#checking the R2
grid_search.best_score_
```

### 4.3 Word Embedding: <a class="anchor" id="subheader43"></a>

As a third potential approach to vectorizing our text, we'll employ word embedding with Gensim and lexvec. This vectorizes the words using a previously trained model. The previously trained model already has a sense of which words are similar/disimilar.

```python
#import word vectors to reference
model = gensim.models.KeyedVectors.load_word2vec_format(
    'lexvec.enwiki+newscrawl.300d.W.pos.vectors', binary=False
)
```

```python
#borrowing this code from representation learning module with BrainStation

def sentence2vec(text):
    
    """
    Embed a sentence by averaging the word vectors of the tokenized text. 
    Out-of-vocabulary words are replaced by the zero-vector.
    -----
    
    Input: text (string)
    Output: embedding vector (np.array)
    """
    tokenized = simple_preprocess(text)
    
    word_embeddings = [np.zeros(300)]
    
    for word in tokenized:
        # if the word is in the model then embed
        if word in model:
            vector = model[word]
        # add zeros for out-of-vocab words
        else:
            vector = np.zeros(300)
            
        word_embeddings.append(vector)
    
    # average the word vectors
    sentence_embedding = np.stack(word_embeddings).mean(axis=0)
     
    return sentence_embedding
```

The above code takes the paragraph, tokenizes the words, and takes the mean of the vectors. Each final vector will end up being composed of 300 numbers. It's less easily interpretable than the other methods but provides a way to compare paragraph to paragraph.

### 4.5 Stemming the Best Model <a class="anchor" id="subheader45"></a>

Using CountVectorizer and TIFDF Vectorizer, there are clearly some similar words that could benefit from stemming or lemmatization. Given time limitations, we'll try stemming on just the TFIDF model, as it outperformed the CountVectorizer.

```python
# identifying the stemmer
stemmer = nltk.stem.PorterStemmer()
```

```python
#creating a custom tokenizer using the stemmer
#borrowed from BrainStation text notebook

def my_tokenizer(sentence):
    # remove punctuation and set to lower case
    for punctuation_mark in string.punctuation:
        sentence = sentence.replace(punctuation_mark,'').lower()

    # split sentence into words
    listofwords = sentence.split(' ')
    listofstemmed_words = []
    
    # remove stopwords and any tokens that are just empty strings
    for word in listofwords:
        if (not word in ENGLISH_STOP_WORDS) and (word!=''):
            # Stem words
            stemmed_word = stemmer.stem(word)
            listofstemmed_words.append(stemmed_word)

    return listofstemmed_words
```

```python
#creating our pipeline
linreg = Pipeline(
    [
        ("vectorizer", TfidfVectorizer(stop_words = 'english', min_df=1, max_df=5000, max_features=2000, tokenizer=my_tokenizer)),
        ("classifier", LinearRegression()),
    ]
)

#ran the grid search using a series of changes on these paramaters
#started with more extreme options for each paramater and then narrowed down to these 3

parameters = {
    "vectorizer__min_df": [9], 
    "vectorizer__max_df": [3000], 
    "vectorizer__max_features": [625],
    "vectorizer__stop_words": ['english'],
    "vectorizer__tokenizer": [my_tokenizer]
}
```

```python
#running the serach to find the best combination of these parameters
grid_search = GridSearchCV(linreg, parameters)

grid_search.fit(X_remain['p1'], y_remain)

print("Best parameters: ", grid_search.best_params_)

#checking the R2
grid_search.best_score_
```
